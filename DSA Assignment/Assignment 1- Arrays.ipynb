```python
def question1(nums,target):
    ans_dict={} #dictionary used to store every value and its index
    for i in range(len(nums)):
        n=target-nums[i]
        if n in ans_dict.keys():
            return [ans_dict[n],i] #if the difference between result and i already in dict we found the pair
        ans_dict[nums[i]]=i
    return []
def input_val():
    nums=list(map(int,input().split()))
    target=int(input())
    res=question1(nums,target)
    print("Output: ",res)
        
input_val()
```

```python
input_val()
```

```python
def question2(nums, val):
    k = 0  # pointer for the last valid index
    for i in range(len(nums)):
        if nums[i] != val:
            nums[k] = nums[i]
            k += 1
    return k
def input_val2():
    nums = list(map(int,input().split()))
    val = int(input())
    result = question2(nums, val)
    print("Output:", result, "nums =", nums[:result] + ["_"] * (len(nums) - result))
input_val2()
```

```python
##In this approach here i'm using binary search to get the time complexity of O(log n)
def question3(nums, target):
    left = 0
    right = len(nums) - 1

    while left <= right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return left
def input_val3():
    nums=list(map(int,input().split()))
    target=int(input())
    res=question3(nums,target)
    print("Output: ",res)
        
input_val3()
```

```python
def question4(digits):
    n = len(digits)
    for i in range(n-1,-1,-1):
        if digits[i]<9:
            digits[i]+=1
            return digits
        digits[i]=0
    return [1]+[0]*len(digits) # for single digit 9
def input_val4():
    digits=list(map(int,input().split()))
    print(question4(digits))
input_val4()
```

```python
def question5(nums1, m, nums2, n):
    p1 = m - 1  # Pointer for nums1
    p2 = n - 1  # Pointer for nums2
    p = m + n - 1  # Pointer for the merged array

    while p1 >= 0 and p2 >= 0:
        if nums1[p1] > nums2[p2]:
            nums1[p] = nums1[p1]
            p1 -= 1
        else:
            nums1[p] = nums2[p2]
            p2 -= 1
        p -= 1

    # Copy any remaining elements from nums2 to nums1
    nums1[:p2 + 1] = nums2[:p2 + 1]

    return nums1
def input_val5():
    nums1=list(map(int,input().split()))
    m=int(input())
    nums2=list(map(int,input().split()))
    n=int(input())
    res=question5(nums1,m,nums2,n)
    print(res)
input_val5()
```

```python
def question7(nums):
    n=nums.count(0)
    nums=[i for i in nums if i!=0]
    nums.extend([0]*n)
    return nums
def input_val7():
    nums=list(map(int,input().split()))
    print(question7(nums))
input_val7()
            
```

```python
def question8(nums):
    duplicate = 0
    numSet=set()
    for num in nums:
        if num in numSet:
            duplicate = num
        else:
            numSet.add(num)
    missing = 0
    for i in range(1, len(nums) + 1):
        if i not in numSet:
            missing = i
    return [duplicate, missing]
def input_val8():
    nums=list(map(int,input().split()))
    print(question8(nums))
input_val8()
    
```
